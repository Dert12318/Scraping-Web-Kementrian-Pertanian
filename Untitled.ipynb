```python
#import package first
#pip install bs4
#pip install selenium
#pip install time
#pip install pandas
#pip install psycopgg2
#pip install os

#import package
import os
import pandas as pd
import time as t
import datetime
import psycopg2
import shutil
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
```

```python
#env
url = "https://aplikasi2.pertanian.go.id/bdsp/id/komoditas"
path_webdriver = "C:/Users/hrd/OneDrive/Documents/File Evan Roberts T/TA/Object Detection/Project-1/WebDriver/chromedriver.exe" #location WebDriver
path_project = r"C:\Users\hrd\OneDrive\Documents\File Evan Roberts T\TA\Object Detection\Project-1" # location Project
delay = 5   # seconds
delay_downloads = 45  # seconds

#Variable
id_subsektor = "subsektor"
id_indikator = "indikator"
id_level = "level"
id_provinsi = "prov"
id_kabupaten = "kab"
id_tahunAwal = "tahunAwal"
id_tahunAkhir = "tahunAkhir"
path_folder_name = r"\downloads"
path_folder_name2 = r"\scraping"


# cronjob
day_to_run = 31
month_to_run = 7
hour_to_run = 20
minute_to_run = 5

#Variable Constant
level = 'Kabupaten'   # selalu kabupaten

#logic


while 1:
    date = datetime.datetime.now()
    if date.day == day_to_run and date.month == month_to_run and date.hour == hour_to_run and date.minute == minute_to_run :
        
        folder_downloads = path_project + path_folder_name
        if not os.path.exists(folder_downloads):
            os.makedirs(folder_downloads)
        #seting download file
        chromeOptions = webdriver.ChromeOptions()
        prefs = {"profile.default_content_settings.popups": 0, "download.default_directory": folder_downloads, "directory_upgrade": True}
        chromeOptions.add_experimental_option("prefs",prefs)
        chromeOptions.add_argument('headless')
        
        #open the website
        driver=webdriver.Chrome(executable_path= path_webdriver,  chrome_options=chromeOptions)
#         driver=webdriver.Chrome(executable_path= path_webdriver)
        driver.get(url)

        #scraping

        #Menu 1
        menu_selector1 = Select(driver.find_element_by_id(id_subsektor))
        menu_options1 = menu_selector1.options
        delay_element1 = WebDriverWait(driver, 10).until(EC.element_to_be_selected(menu_selector1.options[0]))
        #Loop 1
        for index in range(1, len(menu_selector1.options)):
            
            #create folder to download spesific folder
            namefolder = menu_options1[index].text
            path_file = path_project + path_folder_name2
            if not os.path.exists(path_file):
                os.makedirs(path_file)
            location_file1 = os.path.join(path_file, namefolder)
            if not os.path.exists(location_file1):
                os.mkdir(location_file1)
            t.sleep(delay)
            
            option1 = menu_options1[index].get_attribute('value')
            print(option1)
            menu_selector1.select_by_value(option1)
            t.sleep(delay)

            #Menu 2
            menu_selector2 = Select(driver.find_element_by_id(id_indikator))
            menu_options2 = menu_selector2.options
            #delay_element2 = WebDriverWait(driver, 10).until(EC.element_to_be_selected(menu_selector2.options[0]))
            t.sleep(delay)
            #Loop 2
            for index2 in range(1, len(menu_selector2.options)):
                option2 = menu_options2[index2].get_attribute('value')
                print(option2)
                menu_selector2.select_by_value(option2)
                t.sleep(delay)

                #Menu 3
                menu_selector3 = Select(driver.find_element_by_id(id_level))
                print(level) # option3 = level
                menu_selector3.select_by_visible_text(level)
                t.sleep(delay)

                #Menu 4
                menu_selector4 = Select(driver.find_element_by_id(id_provinsi))
                menu_options4 = menu_selector4.options
                t.sleep(delay)
                #delay_element4 = WebDriverWait(driver, 10).until(EC.element_to_be_selected(menu_selector4.options[0]))
                #Loop 3
                for index3 in range(1, len(menu_selector4.options)):
                    option4 = menu_options4[index3].get_attribute('value')
                    print(option4)
                    menu_selector4.select_by_value(option4)
                    t.sleep(delay)

                    #Menu 5
                    menu_selector5 = Select(driver.find_element_by_id(id_kabupaten))
                    menu_options5 = menu_selector5.options
                    t.sleep(delay)
                    #delay_element5 = WebDriverWait(driver, 10).until(EC.element_to_be_selected(menu_selector5.options[0]))
                    #loop 4
                    for index4 in range(1, len(menu_selector5.options)):
                        option5 = menu_options5[index4].get_attribute('value')
                        print(option5)
                        menu_selector5.select_by_value(option5)
                        t.sleep(delay)

                        #Nama File
                        
                        Nama_File = option1 + "_" + option2 + "_" + option4 + "_" + option5 + ".xls" 
                        print(Nama_File)
                        
                        #Menu 6
                        menu_selector6 = Select(driver.find_element_by_id(id_tahunAwal))
                        menu_options6 = menu_selector6.options
                        option6 = menu_options6[len(menu_selector6.options)-1].text # karena indeks 1 adalah tahun yang paling awal ada 1970
                        menu_selector6.select_by_visible_text(option6)
                        t.sleep(delay)

                        #Menu 7
                        menu_selector7 = Select(driver.find_element_by_id(id_tahunAkhir))
                        menu_options7 = menu_selector7.options
                        option7 = menu_options7[len(menu_selector7.options)-1].text # memilih Tahun yang paling akhir
                        menu_selector7.select_by_visible_text(option7)
                        t.sleep(delay)
                        
                        #Download in excel
                        driver.find_element_by_id('excel').click()
                        t.sleep(delay_downloads)
                        
                        #Scarping
                        move_downloads = folder_downloads + r"\DataKeluaranKomoditas.xls"
                        destination = os.path.join(location_file1, Nama_File)
                        t.sleep(delay)
                        shutil.move(move_downloads, destination)
                        option5 = menu_options5[0].get_attribute('value')
                        print(option5)
                        menu_selector5.select_by_value(option5)
                        t.sleep(delay)
                    option4 = menu_options4[0].get_attribute('value')
                    print(option4)
                    menu_selector4.select_by_value(option4)
                    t.sleep(delay)
       
        #exit chrome
        driver.quit()
                
        
        
        
    
    
```
